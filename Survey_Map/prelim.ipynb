```python
df_copy = df
koperasi = ['BUMD', 'Koperasi']
perusahaan = ['PT', 'CV', 'Perorangan']
def label_badan_usaha(val):
    if val in koperasi:
        return 'Koperasi'
    else:
        return 'Perusahaan'

df_copy['Badan Usaha'] = df_copy['Badan Usaha'].apply(label_badan_usaha)
```

```python
grouping0 = df_copy.groupby(['Tipe SPBUN', 'Badan Usaha'])['No'].count().reset_index()
grouping1 = df_copy.groupby(['Tipe SPBUN', 'Badan Usaha', 'Pelabuhan (Pelabuhan & Non Pelabuhan)'])['No'].count().reset_index()
```

```python
conditions = [
    grouping0["Badan Usaha"] == "Koperasi",
    grouping0["Tipe SPBUN"] == "SPBUN Type A",
    grouping0["Tipe SPBUN"] == "SPBUN Type B"
]

choice = [
    grouping0["No"] * .5,
    grouping0["No"] * .5,
    grouping0["No"] * .5
]

grouping0["categorized"] = np.select(conditions, choice)
grouping0
```

```python
# Read the Excel file
excel_path = "/Users/athamawardi/Desktop/Research-Projects/PSE_Pertamina/Survey_Map/lokasi_survei.xlsx"
df = pd.read_excel(excel_path, sheet_name='Calon-Survei')

# Parse coordinates from the "Koordinat" column
def parse_coordinates(coord_str):
    """Parse coordinates from string format 'lat,lon' to [lat, lon]"""
    if pd.isna(coord_str):
        return None
    try:
        parts = str(coord_str).strip().split(',')
        if len(parts) == 2:
            lat = float(parts[0].strip())
            lon = float(parts[1].strip())
            # Validate coordinates are reasonable for Indonesia
            if -15 <= lat <= 10 and 95 <= lon <= 145:
                return [lat, lon]
    except:
        pass
    return None

# Extract coordinates
df['coords'] = df['Koordinat'].apply(parse_coordinates)

# Filter out rows without valid coordinates
df_valid = df[df['coords'].notna()].copy()

print(f"Total rows: {len(df)}")
print(f"Rows with valid coordinates: {len(df_valid)}")
```
